```python
import pygame as pg
import pygame.mixer as mx
import random as rd
import sys

pg.init()

width=600  
height=600  
grids=9
sqrsize=width//grids 
BLACK=(0, 0, 0)
WHITE=(255, 255, 255)

window=pg.display.set_mode((width, height))
pg.display.set_caption("Sudoku-Aubree Thompson")

mx.init()
mx.music.load('DnB Glitchy.mp3')
mx.music.play(-1)


#main loop
def main():
    selected=None  
    while True:
        for event in pg.event.get():
            if event.type==pg.QUIT:
                pg.quit()
                sys.exit()
                mx.music.stop()
            elif event.type==pg.MOUSEBUTTONDOWN:
                #get the mouse coords
                x,y=pg.mouse.get_pos()
                
                #calc the row and column of the selected cell
                row=y//sqrsize
                col=x//sqrsize
                selected=(row,col)
            elif event.type==pg.KEYDOWN:
                if selected is not None:
                    #get pressed key
                    key=event.unicode
                    #TO CLEAR A CELL ENTER 0 INTO THE CELL   
                    if key.isdigit():
                    #put number in selected cell
                        row, col=selected
                        #make it so user can't erase the initial numbers of the puzzle
                        if (row,col) not in initial:
                            grid[row][col]=int(key)
                            if not isvalid(grid, row, col, key):
                                mx.music.load('windows-error-sound-effect-35894')
                                mx.music.play(1)


        
            window.fill(WHITE)
            #draw empty grid
            drawgrid(grid)
            #highlight the selected cell
            if selected is not None:
                highlight(selected)
            pg.display.update()

#function to highlight the selected cell
def highlight(cell):
    row,col=cell
    pg.draw.rect(window,(255, 255, 0),(col*sqrsize,row*sqrsize,sqrsize,sqrsize),3)


def fill(grid):
    #random fill when initializing game
    initial=[]
    for _ in range(20):  
        row=rd.randint(0, 8)
        col=rd.randint(0, 8)
        num=rd.randint(1, 9)
        if grid[row][col]==0 and isvalid(grid, row, col, num):  #check if cell is empty and move is valid
            grid[row][col]=num
            initial.append((row, col)) 
    return initial
            
def isvalid(grid, row, col, num):
        #check if the number already exists in the row or column
        for i in range(9):
            if grid[row][i]==num and i!=col:
                return False
            if grid[i][col]==num and i!=row:  
                return False
        
        #check if the number is alreadyin the subgrids
        startrow,startcol=3*(row//3),3*(col//3)
        for i in range(3):
            for j in range(3):
                if grid[startrow+i][startcol+j]==num and (startrow+i, startcol+j)!=(row, col):
                    return False
        
        return True


def drawgrid(grid):
    font=pg.font.Font(None,36) 

    for i in range(grids):
        for j in range(grids):
            x=j*sqrsize+sqrsize//2-10
            y=i*sqrsize+sqrsize//2-10
            
            textcolor=BLACK

            if grid[i][j]!=0:
                if not isvalid(grid, i, j, grid[i][j]):
                    textcolor=(255, 0, 0)
                text=font.render(str(grid[i][j]),True,textcolor)
                window.blit(text,(x,y))

    for i in range(grids):
        if i%3==0 and i!=0:
            linecolor=BLACK #thicker/black lines for subgribs
            linewdt=2
        else:
            linecolor=(135,206,250) #main thin/blue lines
            linewdt=2

        pg.draw.line(window,linecolor,(0,i*sqrsize),(width,i*sqrsize),linewdt)  
        pg.draw.line(window,linecolor,(i*sqrsize,0),(i*sqrsize,height),linewdt)  

        if i%3==0 and i!=0:
            for j in range(1,3):
                pg.draw.line(window,linecolor,(0,(i+j)*sqrsize),(width,(i+j)*sqrsize),1)  
                pg.draw.line(window,linecolor,((i+j)*sqrsize,0),((i+j)*sqrsize,height),1) 



grid=[[0 for _ in range(9)] for _ in range(9)]

#fill in numbers randomly
initial=fill(grid)

#run main 
if __name__=="__main__":
    main()

```
